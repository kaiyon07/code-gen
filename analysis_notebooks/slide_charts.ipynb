```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the data
file_path = "filtered_metrics_results_metadata.csv"  # Update this path if necessary
df = pd.read_csv(file_path)

# Correct the naming: Rename V1 to V2 and V2 to V1
def correct_version(file_name):
    if "PROMPT_REFACTOR_V1" in file_name:
        return "V2"  # Previously V1 is now V2
    elif "PROMPT_REFACTOR_V2" in file_name:
        return "V1"  # Previously V2 is now V1
    elif "PROMPT_REFACTOR_V3" in file_name:
        return "V3"
    else:
        return None

df["Version"] = df["File"].apply(correct_version)
df_filtered = df[df["Version"].notna()]

# Extract model names
df_filtered["Model"] = df_filtered["File"].str.extract(r"^(.*?)/PROMPT_REFACTOR")[0]

# Create a mapping for ordered names
model_name_mapping = {
    "gemini_results_temp_0/gemini-1.5-flash": "Gemini 1.5 Flash",
    "gemini_results_temp_0/gemini-1.5-pro": "Gemini 1.5 Pro",
    "llama3_90B_results_temp_0": "LLaMa3-90B",
    "mixtral_results_temp_0": "Mixtral",
    "openai_results_temp_0/gpt-3.5-turbo": "GPT 3.5 Turbo",
    "openai_results_temp_0/gpt-4-turbo": "GPT 4 Turbo"
}
df_filtered["Model"] = df_filtered["Model"].map(model_name_mapping)

# Compute average accuracy and linting errors per model and version
grouped = df_filtered.groupby(["Model", "Version"])[
    ["Average Accuracy (Refactored)", "Average ESLint Change Percentage", "Average Closure Change Percentage"]
].mean().reset_index()

# Define the desired order for the legend
desired_order = [
    "Gemini 1.5 Flash",
    "Gemini 1.5 Pro",
    "LLaMa3-90B",
    "Mixtral",
    "GPT 3.5 Turbo",
    "GPT 4 Turbo"
]

# Ensure output directory and subfolder exist
output_dir = os.path.join("slide_visuals", "metadata")
os.makedirs(output_dir, exist_ok=True)

# Create the scatter plot with larger dots
plt.figure(figsize=(12, 8))

# Unique colors for each model
model_colors = {name: plt.cm.tab10(i % 10) for i, name in enumerate(desired_order)}

for model in desired_order:
    model_data = grouped[grouped["Model"] == model]
    # Plot ESLint as circles
    for _, row in model_data.iterrows():
        plt.scatter(
            row["Average Accuracy (Refactored)"],
            row["Average ESLint Change Percentage"],
            label=model if model not in plt.gca().get_legend_handles_labels()[1] else "",
            s=150,
            color=model_colors[model],
            marker='o',  # Circle for ESLint
        )
        # Annotate the point with the version
        plt.text(
            row["Average Accuracy (Refactored)"] - 0.01,
            row["Average ESLint Change Percentage"],
            row["Version"],
            fontsize=10,
            color="black",
        )
    # Plot Closure as stars
    for _, row in model_data.iterrows():
        plt.scatter(
            row["Average Accuracy (Refactored)"],
            row["Average Closure Change Percentage"],
            label=None,  # No additional label for the same model
            s=200,
            color=model_colors[model],
            marker='*',  # Star for Closure
        )
        # Annotate the point with the version
        plt.text(
            row["Average Accuracy (Refactored)"] - 0.01,
            row["Average Closure Change Percentage"],
            row["Version"],
            fontsize=10,
            color="black",
        )

# Update legend with the desired order
handles, labels = plt.gca().get_legend_handles_labels()
ordered_handles = [handles[labels.index(name)] for name in desired_order if name in labels]
plt.legend(ordered_handles, desired_order, loc="upper right", title="Models", fontsize=10)

# Add title, labels, and grid
plt.title("Model-level Comparison of Accuracy and Linting Errors (V1, V2, V3)", fontsize=14)
plt.xlabel("Average Accuracy (Refactored)", fontsize=12)
plt.ylabel("Linting Errors (% Change)", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()

# Save the scatter plot in the metadata subfolder
scatter_plot_path = os.path.join(output_dir, "scatter_accuracy_and_linting.png")
plt.savefig(scatter_plot_path)
plt.show()

print(f"Scatter plot saved at {scatter_plot_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load the data
file_path = "filtered_metrics_linter_outputs.csv"  # Update this path if necessary
df = pd.read_csv(file_path)

# Filter relevant data for V1 and V4
versions = ["PROMPT_REFACTOR_V1", "PROMPT_REFACTOR_V4"]
df_filtered = df[df["File"].str.contains('|'.join(versions))]

# Extract model names and versions
df_filtered["Version"] = df_filtered["File"].apply(
    lambda x: "V1" if "PROMPT_REFACTOR_V1" in x else
              "V4"
)
df_filtered["Model"] = df_filtered["File"].str.extract(r"^(.*?)/PROMPT_REFACTOR")[0]

# Create a mapping for ordered names
model_name_mapping = {
    "gemini_results_temp_0/gemini-1.5-flash": "Gemini 1.5 Flash",
    "gemini_results_temp_0/gemini-1.5-pro": "Gemini 1.5 Pro",
    "llama3_90B_results_temp_0": "LLaMa3-90B",
    "mixtral_results_temp_0": "Mixtral",
    "openai_results_temp_0/gpt-3.5-turbo": "GPT 3.5 Turbo",
    "openai_results_temp_0/gpt-4-turbo": "GPT 4 Turbo"
}
df_filtered["Model"] = df_filtered["Model"].map(model_name_mapping)

# Compute average accuracy and linting errors per model and version
grouped = df_filtered.groupby(["Model", "Version"])[
    ["Average Accuracy (Refactored)", "Average ESLint Change Percentage", "Average Closure Change Percentage"]
].mean().reset_index()

# Define the desired order for the legend
desired_order = [
    "Gemini 1.5 Flash",
    "Gemini 1.5 Pro",
    "LLaMa3-90B",
    "Mixtral",
    "GPT 3.5 Turbo",
    "GPT 4 Turbo"
]

# Create the scatter plot with larger dots
plt.figure(figsize=(12, 8))

# Unique colors for each model
model_colors = {name: plt.cm.tab10(i % 10) for i, name in enumerate(desired_order)}

for model in desired_order:
    model_data = grouped[grouped["Model"] == model]
    # Plot ESLint as circles
    for _, row in model_data.iterrows():
        plt.scatter(
            row["Average Accuracy (Refactored)"],
            row["Average ESLint Change Percentage"],
            label=model if model not in plt.gca().get_legend_handles_labels()[1] else "",
            s=150,
            color=model_colors[model],
            marker='o',  # Circle for ESLint
        )
        # Annotate the point with the version
        plt.text(
            row["Average Accuracy (Refactored)"] - 0.01,
            row["Average ESLint Change Percentage"],
            row["Version"],
            fontsize=10,
            color="black",
        )
    # Plot Closure as stars
    for _, row in model_data.iterrows():
        plt.scatter(
            row["Average Accuracy (Refactored)"],
            row["Average Closure Change Percentage"],
            label=None,  # No additional label for the same model
            s=200,
            color=model_colors[model],
            marker='*',  # Star for Closure
        )
        # Annotate the point with the version
        plt.text(
            row["Average Accuracy (Refactored)"] - 0.01,
            row["Average Closure Change Percentage"],
            row["Version"],
            fontsize=10,
            color="black",
        )

# Update legend with the desired order
handles, labels = plt.gca().get_legend_handles_labels()
ordered_handles = [handles[labels.index(name)] for name in desired_order if name in labels]
plt.legend(ordered_handles, desired_order, loc="upper right", title="Models", fontsize=10)

# Add title, labels, and grid
plt.title("Model-level Comparison of Accuracy and Linting Errors (V1, V4)", fontsize=14)
plt.xlabel("Average Accuracy (Refactored)", fontsize=12)
plt.ylabel("Linting Errors (% Change)", fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()

# Ensure output directory exists
output_dir = os.path.join("slide_visuals", "linter_outputs")
os.makedirs(output_dir, exist_ok=True)

# Save the scatter plot with the updated file name
scatter_plot_path = os.path.join(output_dir, "scatter_accuracy_and_linting.png")
plt.savefig(scatter_plot_path)
plt.close()

print(f"Scatter plot saved at {scatter_plot_path}")
```
